# Ventas por tipo de producto de Cuidado personal

```python
producto = 'Cuidado personal'
filtro_CuidadoPersonal = datos_ventas[datos_ventas['Tipo de producto'] == producto]
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Zona', data=filtro_CuidadoPersonal)
plt.title('Distribución por zonas')
plt.show()
```

# Ventas Online

```python
Canal = 'Online'
canal_ventas = datos_ventas[datos_ventas['Canal de venta'] == Canal]
canal_ventas
```

```python
canal_ventas.shape[0]
```

```python
recaudacion_por_zona = canal_ventas.groupby('Zona')['Importe venta total'].sum().reset_index()
plt.figure(figsize=(15, 7))
sns.barplot(x='Zona', y='Importe venta total', data=recaudacion_por_zona, order=recaudacion_por_zona.sort_values('Importe venta total', ascending=False)['Zona'])
plt.title('Recaudación por Continente')
plt.xlabel('Zona')
plt.ylabel('Suma del Importe de Venta Total')
plt.show()
```

```python
recaudacion_por_zona
```

# Total de Ventas Online

```python
total_VOnline = canal_ventas['Importe venta total'].sum()
```

```python
total_VOnline
```

# Ventas Offline

```python
canal_ventasOff = datos_ventas[datos_ventas['Canal de venta'] == 'Offline']
canal_ventasOff
```

```python
recaudacion_offline = canal_ventasOff.groupby('Zona')['Importe venta total'].sum().reset_index()
plt.figure(figsize=(15, 7))
sns.barplot(x='Zona', y='Importe venta total', data=recaudacion_offline, order=recaudacion_offline.sort_values('Importe venta total', ascending=False)['Zona'])
plt.title('Recaudación por Continente Offline')
plt.xlabel('Zona')
plt.ylabel('Suma del Importe de Venta Total')
plt.show()
```

```python
recaudacion_offline
```

# Total de ventas Offline

```python
total_VOffline = recaudacion_offline['Importe venta total'].sum()
```
